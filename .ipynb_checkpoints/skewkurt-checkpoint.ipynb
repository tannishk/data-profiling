# Skewness and Kurtosis Module
This module is used to calculate skewness and kutosis module 

```python
import pandas as pd
import skewkurt as sk
a = pd.read_csv('sample data.csv')
a.head()
```

# skewness(dataset,col):
To Calculate the skewness of a given data For normal distribute data skewness = 0 , Skewness > 0 more weight the left tail and less weight in right tail

```python
sk.skewness(a,'Income')
```

### kurtosis(dataset,col,ty):
To calculate Kurtosis of a given data . It can be used to calculate both fisher and pearson projection.  ty can be fisher or pearson 

TO CALUATE FISHER KURTOSIS

```python
sk.kurtosis(a,'Income','fisher')
```

TO CALCULATE PEARSON KURTOSIS

```python
 sk.kurtosis(a,'Income','fisher')
```
