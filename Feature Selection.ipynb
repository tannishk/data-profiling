# Feature Selection 
This module is generally used for feature selection and feature transformation which helps us to overcome the problem of curse of dimentionality and improve estimators’ accuracy scores or to boost their performance on very high-dimensional datasets

```python
import pandas as pd
import featureselection as fs
import muloutlier as ml
a = pd.read_csv('sample data.csv')
```

### varthres(dataset,threshold)
To select the columns based on the value of their varience which is greater then the threhold but only numeric dataset should be passed to this function . It will return the altered dataset in the form numpy array 

```python
for i in a:
    ##print i
    a = ml.transform(a,i) ## To transform textual data into numeric form 
a.head()
```

```python
fs.varthres(a,0.5)
```

### Kbest(df,features,class1,arr):
It is used to select arr number of best features based on chi test compared between features and  class1 which indicate the class to be predicted . Here df is the dataframe used . It only accepts features in numeric form . It returns the altered dataset in numpy array form and features which are selected  .

```python
fs.Kbest(a,['State','Gender','Income','Location Code','Policy'],'Vehicle Class',4)
```

### Pca(dataset,features,numberofcomponenets):
It is used to apply the Principle component analysis for the given dataset with these features transforming it into data with number of componenets 
Output : A transformed output numpy array

```python
fs.Pca(a,['State','Gender','Income','Location Code','Policy'],4)
```

### Ica(dataset)
To do feature transformation such as all features are independent from one another 
Input : Dataset is the input 
output : Transformed dataset in the form of numpy array where features are independent from one another 

```python
fs.Ica(a)
```
